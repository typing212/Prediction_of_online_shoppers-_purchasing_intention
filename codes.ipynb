**Accumulation time period**: 1 year

**Numerical attributes**
1. "Administrative" - Number of pages visited for administrative purposes.
2. "Administrative Duration" - Time spent (in seconds) by visitor on administrative pages.
3. "Informational" - Number of pages visited for informational purposes.
4. "Informational Duration" - Time spent (in seconds) by visitor on informational pages.
5. "Product Related" - Number of pages visited that are product related.
6. "Product Related Duration" - Time spent (in seconds) by visitor on product related pages.
7. “Bounce rate” – Average bounce rate of pages visited by visitor. Bounce rate of a page 
refers to the percentage of visitors who enter and leave the website from that page without 
triggering any other requests to the Google Analytics server during that session.
8. "Exit Rate" – Average exit rate of pages visited by visitor. Exit rate of a page refers to the 
percentage of page views to that page which was the last in a session.
9. "Page Value" – Average page value of pages visited by visitor. Page value of a page refers 
to the average value of the page that a user visited before completing an e-commerce 
transaction.
10. "Special Day" – Value that indicates the closeness of the visit session to a specific special 
day (e.g. Valentine's Day). The value of this attribute is determined by considering the 
dynamics of e-commerce such as the duration between the order date and delivery date. 
For example, for Valentine’s day, this value is non-zero between February 2 and February 
12, and has a value of zero before and after this time period. It reaches its maximum value 
of 1 on February 8.

**Categorical attributes**
11. “Operating system” – Operating system of visitor.
12. “Browser” – Internet browser of visitor.
13. “Region” – Geographic region of session.
14. “Traffic type” – Source by which visitor arrived at website (e.g. banner, direct, SMS).
15. “Visitor type” – Visitor type as “New Visitor,’’ ‘‘Returning Visitor,’’ and ‘‘Other’’.
16. “Weekend” - Boolean value indicating whether visit date is on a weekend.
17. “Month” – Month of visit date.
18. “Revenue” – Class label indicating whether visit was finalised with a transaction

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, f1_score, roc_auc_score
from sklearn.feature_selection import mutual_info_classif
from mrmr import mrmr_classif

import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset

from imblearn.under_sampling import RandomUnderSampler
from imblearn.over_sampling import RandomOverSampler
from imblearn.over_sampling import SMOTE
```

```python
def preprocess_categorical(df):
    categorical = ['OperatingSystems','Browser','Region','TrafficType','VisitorType','Month']
    
    dummies = pd.get_dummies(df[categorical].astype(str))
    df = df.drop(categorical,1)
    df = pd.concat([df,dummies],axis =1)

    return df

def preprocess_numerical(train,test):
    numerical = ['Administrative', 'Administrative_Duration', 'Informational',
               'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration',
               'BounceRates', 'ExitRates', 'PageValues', 'SpecialDay']
    for i in numerical:
        mean = train[i].mean()
        std = train[i].std()
        train[i] = (train[i]-mean)/std
        test[i] = (test[i]-mean)/std
        
    return train,test
```

```python
class MLP(nn.Module):
    '''
    Multilayer Perceptron.
    '''
    def __init__(self, input_dim, hidden_dim):
        super().__init__()
        self.layers = nn.Sequential(
            nn.Linear(input_dim,hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim,1),
            nn.Sigmoid())


    def forward(self, x):
        '''Forward pass'''
        return self.layers(x).reshape(-1,)
  
  

batch_size = 16
epochs = 20 # use 10 for models with feature selection, use 20 for others

# Set fixed random number seed
torch.manual_seed(42)

# Prepare dataset
X_train_tensor = torch.tensor(X_train.values,dtype = torch.float)
y_train_tensor = torch.tensor(y_train.values,dtype = torch.float)
train = TensorDataset(X_train_tensor, y_train_tensor)
train_dataloader = DataLoader(train, batch_size=batch_size, shuffle=True, drop_last=True)

X_test_tensor = torch.tensor(X_test.values,dtype = torch.float)
y_test_tensor = torch.tensor(y_test.values,dtype = torch.float)
test = TensorDataset(X_test_tensor, y_test_tensor)
test_dataloader = DataLoader(test, batch_size=batch_size, shuffle=True, drop_last=True)
```

## MLP with different number of neurons in hidden layer using full features

```python
# MLP with different number of neurons in hidden layer using full features

Hidden = [10,20,40]

for hidden_dim in Hidden:
    print('------------------------------------')
    print('Hidden_dim =',hidden_dim)
    print('------------------------------------')
    # Initialize the MLP
    mlp = MLP(input_dim = 74, hidden_dim = hidden_dim)
    # Define the loss function and optimizer
    loss_function = nn.BCELoss()
    optimizer = torch.optim.Adam(mlp.parameters())
    
    train_loss = []
    test_loss = []
    train_acc = []
    test_acc = []
    # Run the training loop
    for epoch in range(epochs):

        running_loss = 0.0
        predss = torch.tensor([])  
        targetss = torch.tensor([])
        for batch_idx, (x, targets) in enumerate(train_dataloader):

            optimizer.zero_grad()
            pred = mlp(x)
            loss = loss_function(pred, targets)

            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            predss = torch.cat((predss,pred),0)
            targetss = torch.cat((targetss,targets),0)
            
        train_loss.append(running_loss/(batch_idx+1))
        
        cm = confusion_matrix(targetss.int().numpy(), torch.round(predss).int().numpy(), labels=[1, 0], sample_weight=None) # confusion matrix 
        TP = cm[0][0] # True Positives                                                                                          
        FN = cm[0][1] # False Positives                                                                                          
        FP = cm[1][0] # True Negatives                                                                                          
        TN = cm[1][1] # False Negatives  
        Accuracy = (TP+TN)/(TP+FN+FP+TN)
        
        train_acc.append(Accuracy)
        print('epoch {:3d} | Train loss: {:.4f} | Train acc: {:.4f}'.format(epoch, running_loss/(batch_idx+1), Accuracy))

        vpred = torch.tensor([])  
        vtargets = torch.tensor([])
        vrunning_loss = 0.0
        with torch.no_grad():
            for j, (vx,vtarget) in enumerate(test_dataloader, 0):

                vpred = torch.cat((vpred,mlp(vx)),0)
                vtargets = torch.cat((vtargets,vtarget),0)

                vloss = loss_function(vpred, vtargets)
                vrunning_loss += vloss.item()
            
        test_loss.append(vrunning_loss/(j+1))
        cm = confusion_matrix(vtargets.int().numpy(), torch.round(vpred).int().numpy(), labels=[1, 0], sample_weight=None) # confusion matrix 

        TP = cm[0][0] # True Positives                                                                                          
        FN = cm[0][1] # False Positives                                                                                          
        FP = cm[1][0] # True Negatives                                                                                          
        TN = cm[1][1] # False Negatives  

        TPR = TP/(TP+FN)
        TNR = TN/(TN+FP) # True Negative Rate (specificity)  
        Accuracy = (TP+TN)/(TP+FN+FP+TN)

        f1score = f1_score(vtargets.int().numpy(), torch.round(vpred).int().numpy())
        AUC = roc_auc_score(vtargets.int().numpy(), vpred.numpy())
        test_acc.append(Accuracy)
        print('epoch {:3d} | Test Accuracy: {:.4f} | Test TPR: {:.4f} | Test TNR: {:.4f} | Test F1 Score: {:.4f} | Test AUC: {:.4f}'.format(epoch, Accuracy, TPR, TNR, f1score, AUC))

    print('Finished Training')
    
    plt.plot( list(range(1,len(train_loss)+1)), train_loss, label='Train Loss')
    plt.plot( list(range(1,len(train_loss)+1)), test_loss, label='Test Loss')

    plt.xlabel('Number of epochs')
    plt.ylabel('BCE Loss')
    plt.legend()
    plt.savefig('loss_hidden_dim ' + str(hidden_dim) + '.png')
    plt.show()
    
    plt.plot( list(range(1,len(train_loss)+1)), train_acc, label='Train Accuracy')
    plt.plot( list(range(1,len(train_loss)+1)), test_acc, label='Test Accuracy')

    plt.xlabel('Number of epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.savefig('acc_hidden_dim ' + str(hidden_dim) + '.png')
    plt.show()
```

```python
def recall_10(y_true, y_pred):
    recall_10 = metrics.recall_score(y_true, y_pred,labels=[1,0])
    return recall_10

score = make_scorer(recall_10, greater_is_better=True)

def dt_model(X_train, y_train, model_name):
    
    dt = DecisionTreeClassifier()
    parameters = {'criterion':['gini','entropy'],'max_depth':[10,50,100,300]}
 
    grid_search = GridSearchCV(dt,parameters,n_jobs=-1,verbose=1,scoring=score)
    grid_search.fit(X_train,y_train)
    best_score = grid_search.best_score_
    best_parameters = grid_search.best_estimator_.get_params()
    print('best_score:', best_score)
    print('best_parameters:', best_parameters)
    
    dt = DecisionTreeClassifier(criterion=best_parameters['criterion'], max_depth=best_parameters['max_depth'])
    dt.fit(X_train,y_train)
    y_train_pred = dt.predict(X_train)
    y_test_pred = dt.predict(X_test)

    y_test_pred_proba = dt.predict_proba(X_test)
#     print("Train Metrics: ", metrics.classification_report(y_train, y_train_pred))
    print("Test Metrics: ", metrics.classification_report(y_test, y_test_pred))
    
    acc = metrics.accuracy_score(y_test, y_test_pred)
    tpr = metrics.recall_score(y_test, y_test_pred,labels=[1,0])
    auc = metrics.roc_auc_score(y_test, y_test_pred_proba[:,1],labels=[1,0])
    precision = metrics.precision_score(y_test, y_test_pred,labels=[1,0])
    f1 = metrics.f1_score(y_test, y_test_pred,labels=[1,0])
    
    cm = metrics.confusion_matrix(y_test, y_test_pred,labels=[1,0])
    print("Test Confusion Matrix:\n", cm)
    
    explainer = shap.TreeExplainer(dt)
    shap_values = explainer.shap_values(X_train)

    shap.summary_plot(shap_values, X_train)
    plt.savefig(model_name+'.png')

    return {'acc': acc, 'precision':precision, 'f1': f1, 'cm': cm, 'auc': auc, 'tpr':tpr}
```

```python
import pandas as pd
import numpy as np
import shap
import xgboost as xgb
from xgboost.sklearn import XGBClassifier
from sklearn import metrics
from sklearn.model_selection import GridSearchCV  
from matplotlib.pylab import rcParams
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import roc_auc_score
import matplotlib.pylab as plt
%matplotlib inline

rcParams['figure.figsize'] = 12, 4
pd.set_option('display.max_columns', None)
```

```python
train = pd.read_csv('Train.csv')
test = pd.read_csv('Test.csv')
```

```python
# Define a model-tunning with CV using XGboost

def modelfit(alg, dtrain, predictors, useTrainCV=True, cv_folds=5, early_stopping_rounds=50):
    
    if useTrainCV:
        xgb_param = alg.get_xgb_params()
        xgtrain = xgb.DMatrix(dtrain[predictors].values, label=dtrain[target].values)
        cvresult = xgb.cv(xgb_param, xgtrain, num_boost_round=alg.get_params()['n_estimators'], nfold=cv_folds,
            metrics='auc', early_stopping_rounds=early_stopping_rounds)
        alg.set_params(n_estimators=cvresult.shape[0])
    
    #Fit the algorithm on the data
    alg.fit(dtrain[predictors], dtrain[target],eval_metric='auc')
        
    #Predict training set:
    dtrain_predictions = alg.predict(dtrain[predictors])
    dtrain_predprob = alg.predict_proba(dtrain[predictors])[:,1]
        
    #Print model report:
    print("\nModel Report")
    print("Accuracy : %.4g" % metrics.accuracy_score(dtrain[target].values, dtrain_predictions))
    print("AUC Score : %f" % metrics.roc_auc_score(dtrain[target], dtrain_predprob))
    
    cm = confusion_matrix(dtrain[target].values, dtrain_predictions)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=alg.classes_)
    disp.plot()
    plt.show()
```

```python
# Define target value
target = 'Revenue'

# Choose all predictors except target column
predictors = [x for x in train.columns if x not in [target]]
```
